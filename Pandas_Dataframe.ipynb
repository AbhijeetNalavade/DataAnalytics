```python
# Q1. Creating a DataFrame
# Create a Pandas DataFrame using a Python dictionary with the following columns:

# Product

# Region

# Sales

# Quantity

import pandas as pd
# creating a dictionary of required col
dataset = {
    'Product': ['Laptop', 'Android', 'Apple', 'Headphones', 'Alexa', 'Computer'],
    'Region': ['North', 'North', 'South', 'South', 'North', 'South'],
    'Sales': [100, 200, 150,300,150,340],
    'Quantity':[10,15,20,15,20,10]
}
# Convert the dictionary into a DataFrame
df = pd.DataFrame(dataset)
print(df)

```

```python
'''Add at least 5 rows of sample data and display the DataFrame.

Dataset for Questions 2–10

Use the dataset file provided'''

# read the data
df = pd.read_csv('/content/Retail_dataset.csv')
df.head(10)
```

```python
'''Q2. Data Inspection

Perform the following inspection tasks:

Display the first 5 rows of the dataset.

Display the last 5 rows.

Show the structure of the dataset using info().

Display the shape of the dataset.
'''
print("--- HEAD (First 5 Rows) ---")
print(df.head())

print("\n--- TAIL (Last 5 Rows) ---")
print(df.tail())

print("\n--- INFO (Data Types & Summary) ---")
df.info()  # df.info() prints itself automatically

print("\n--- SHAPE (Rows, Columns) ---")
print(df.shape)
```

```python
'''Q3. Column Selection

Select and display only the following columns:

Product

Region

Sales'''
display_columns = ["Product","Region","Sales"]
selecte_col = df[display_columns]
print(selecte_col)

```

```python
'''Q4. Indexing using .loc
Using .loc, perform the following:

Display the row with index 10.

Display the Sales value of the row with index 20'''

# Display the row with index 10.
display_rows = df.loc[[10]]
print(display_rows)

# Display the Sales value of the row with index 20
sales_data = df.loc[20,"Sales"]
print(f"the sales for row 20 is :{sales_data}")


```

```python
'''Q5. Indexing using .iloc

Using .iloc, display:

The first 6 rows and first 4 columns.

The Sales value from the 3rd row.'''

top_rows_col = df.iloc[0:6,0:4]
print("The first 6 rows and first 4 columns ")
display(top_rows_col)
# Position 2 targets the 3rd physical row
# Wrapping 4 in brackets keeps it as a table structure for display due to iloc
sales_data = df.iloc[2,[4]]
print("The Sales value from the 3rd row")
display(sales_data)
```

```python
'''Q6. Row Slicing

Display:

The first 10 rows of the dataset.

Rows 15 to 25'''

# The first 10 rows of the dataset.
first_10_rows = df.head(10)
display(first_10_rows)

# Rows 15 to 25
rows_15_to_25 = df.iloc[15:26]
display(rows_15_to_25)

```

```python
'''Q7. Filtering Data

Filter the dataset to display:

Rows where Region = "East".

Rows where Sales > 1500.'''
filter_region = df[df["Region"] == "East"]
display(filter_region)

# Rows where Sales > 1500.
filter_sales = df[df["Sales"] > 1500]
display(filter_sales)



```

```python
'''Q8. Multiple Condition Filtering

Display rows where:

Region is West

Sales is greater than 1200'''

region_filter = df[(df['Region'] == 'West') & (df['Sales'] > 1200)]
display(region_filter)

```

```python
'''Q9. Descriptive Statistics

Generate summary statistics using:

describe()

From the result identify:

Mean sales value

Maximum quantity sold

Minimum sales value'''

descriptive_stats = df.describe()
display(descriptive_stats)

# 2. Extract and print the specific values asked in the prompt
mean_sales = descriptive_stats.loc["mean", "Sales"]
max_qty = descriptive_stats.loc["max", "Quantity"]
min_sales = descriptive_stats.loc["min", "Sales"]

print(f"Mean sales value: {mean_sales}")
print(f"Maximum quantity sold: {max_qty}")
print(f"Minimum sales value: {min_sales}")



```

```python
'''Q10. Aggregation Analysis

Using the Sales column, calculate:

Total sales

Average sales

Maximum sales

Minimum sales'''
Aggregation_analysis = df["Sales"]
total_sales = Aggregation_analysis.sum()
average_sales = Aggregation_analysis.mean()
max_sales = Aggregation_analysis.max()
min_sales = Aggregation_analysis.min()

print("total sales for the given data :",total_sales)
print("Average sales for the given data :",average_sales)
print("Maximum sales for the given data :",max_sales)
print("Minimum sales for the given data :",min_sales)
```
